The fundamental goal of portfolio theory is to optimally allocate investments between different assets. Mean variance optimization (MVO) is a quantitative tool which makes this allocation decision by considering the trade-off between risk and return.

Portfolio theory can be though of as two seperate procedures: (1) estimation of the expected return for all securities and the covariances between the securites; and (2) determening the optimal capital to be invested in each security given the estimations from (1).  MVO is a procedure for estimating the optimal weighting of each security in a portfolio.  It maximizes a reward-to-variability ratio to give an investor the greatest return for a given level of risk.  

We provide both an overview of theory and implementation of MVO to make an asset allocation decision.  The tutorial requires the following imports:

```python
# Std. Modules
import datetime
from itertools import product

# Vectors
import numpy as np

# Optimizers
import scipy as sp
from scipy.optimize import minimize

#  Data Manipulation
import pandas as pd
from pandas.io.data import DataReader

# Figures / Graphs
%matplotlib inline
from matplotlib import cm
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.ticker import LinearLocator, FormatStrFormatter
```

<h1>Mean Variance Optimization Framework</h1>

MVO is a mathematical procedure to construct a portfolio.  It requires the maximization of an objective function that considers both risk and return.  The most common objective function is called the Sharpe Ratio and is calculated as follows:

$$
Sharpe Ratio = \frac{RET_P - r_f}{\sigma_P} =
\frac{\sum\limits_{i=1}^N w_i RET_i - r_f}{\sum\limits_{i=1}^N \sum\limits_{j=1}^M w_i w_j COV(RET_i, RET_j)}
$$

Where $RET_P$ is the average daily return of the portfolio, $r_f$ is the risk-free rate, $w_i$ is the weight of security $i$ in the portfolio of $N$ ($M$) securities, and $RET_i$ is the return of security $i$.

For shorthand, we will denote the Sharpe Ratio:, $S_{RET}(w_1, w_2, ..., w_N)$.  The ratio can be thought of as a function with weights as parameters with a given set of expected returns ($RET$).  

Mathematically, MVO is the procedure which maximizes the Sharpe Ratio subject to constraints:

$$ 
\begin{equation*}
 \begin{aligned}
 & \underset{w_1, w_2, ..., w_N}{\text{maximize}}
 & & S_{RET}(w_1, w_2, ..., w_N) \\ \\
 & \text{subject to}
 & & w_i \geq 0, \; \sum w_i = 1 \\
 & & & i = 1, \ldots, N.
 \end{aligned}
\end{equation*}
$$

That is, there is no short selling and all securites must have a positive weight.

<h1>MVO Two Asset Example</h1>

We use a two asset example to demonstrate MVO.  From this example, it will be easy to generalize for an arbitrary number of assets (note, the computational complexity on the computer will increase significantly).  We use two Vanguard ETFs in the example: Total Stock Market ETF (VTI) and Total Bond Market ETF (BND) and assume a risk-free rate ($r_f$) of zero in the current climate (June 2015).

The first part of portfolio theory requires estimating future expected returns and covarainces.  We use a naive approach in the estimation, using historical 1-year average daily returns and covariances.  This method is not appropriate for use when investing real money, but will suffice for demonstration purposes.  We first download the data and calculate the return and covariance matrices:

```python
# Start and end periods
start = datetime.datetime(2014, 1, 1) # 1/1/2014
end = datetime.datetime(2015, 1, 1)   # 1/1/2015

# Download finance data
vti = DataReader('VTI', 'yahoo', start=start, end=end) # Vanguard Total Stock Market ETF (VTI)
bnd = DataReader('BND', 'yahoo', start=start, end=end) # Vanguard Total Bond Market ETF (BND)

# Calculate returns (RET = P2 / P1 - 1)
vti['RET'] = vti['Adj Close'] / vti['Adj Close'].shift(1) - 1
bnd['RET'] = bnd['Adj Close'] / bnd['Adj Close'].shift(1) - 1

# Fill missing values with zero 
vti['RET'].fillna(0, inplace=True)
bnd['RET'].fillna(0, inplace=True)

# Create return and covariance matrix
ret = np.array( zip(vti['RET'].values, bnd['RET'].values) )
cov = np.cov(ret.T)
```

Next, we graph both ETFs.  Notice that the bond ETF (BND) is significantly less volatitle compared to the stock ETF (VTI).  The blue line represents the daily returns of VTI and the green line indicates the daily returns of BND over the period of January 2014 to December 2014.  

```python
# Create figure and axis
fig, ax = plt.subplots()

# Plot VTI and BND returns
ax.plot(vti.index, vti['RET'].values, label='VTI')
ax.plot(vti.index, bnd['RET'].values, label='BND')

# Add title and axis labels
plt.title('1 Year Daily Returns')
ax.set_xlabel('Time')
ax.set_ylabel('Returns')

# Add legend and date formatting
ax.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.0)
fig.autofmt_xdate()

# Draw figure
plt.show()
```

In order to calculate the mean-variance efficient portfolio, we must first define some helper functions in Python:

```python
# Portfolio Weighted-Average Return
def port_avg_ret(wt, ret):
    return  np.mean( np.sum( wt*ret , axis=1) )

# Portfolio Std. Deviation
def port_sd(wt, cov):
    return np.sqrt( np.dot(wt, np.dot(wt, cov)) )

# Portfolio Sharpe Ratio
def sharpe(wt, ret, cov):
    return (port_avg_ret(wt, ret) - 0.0) / port_sd(wt, cov)
```

We now move our attention to optimizing the Sharpe Ratio given the constraints of no short selling.  We can view the asset allocation decision as finding the optimal weights in the portfolio.  The graph below shows the Sharpe Ratio for various weights.

```python
# X and Y axis bounds
lower = -2
upper = 2

#  Create 3D plotting figure and axis
fig = plt.figure()
ax = fig.gca(projection='3d')

# Generate weights and Sharpe Ratios to graph
temp = [np.arange(0.0, 1.01, .01)]*2
wts = np.array([elm for elm in product(*temp) if  sum(elm) == 1])
sharpe_ratios = np.array( [sharpe(w, ret, cov) for w in wts] )

# Add axis labels
ax.set_xlabel(r'$w_{IWV}$', fontsize=20)
ax.set_ylabel(r'$w_{BND}$', fontsize=20)
ax.set_zlabel(r'$S_{RET}(w_{IWV}, w_{BND})$', fontsize=20)

# Format X, Y, and Z axis
ax.xaxis.set_major_formatter(plt.NullFormatter())
ax.yaxis.set_major_formatter(plt.NullFormatter())
ax.zaxis.set_major_formatter(plt.NullFormatter())

# Plot Sharpe Ratio for different combinations of portfolio weights
wt0, wt1 = np.meshgrid(*wts.T)
surf = ax.plot_surface(wt0, wt1, sharpe_ratios, 
                       rstride=1, cstride=1, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# Draw figure
plt.show()
```

We can also view this relationship by observing the various return and standard deviation tradeoffs of the portfolios generated from using different weighting schemes.  The region on the line that have the highest average return for a given level of risk is known as the efficient frontier. 

```python
# Efficient Frontier
means = [port_avg_ret(w, ret) for w in wts]
stds = [port_sd(w, cov) for w in wts]
plt.plot(stds, means)
plt.xlabel('Std. Deviation')
plt.ylabel('Avg. Return')
plt.title('Efficient Frontier')
plt.show()
```

<h1>Optimization</h1>

Finally, we turn our attention to maximizing the Sharpe Ratio.  To accomplish this goal, we use Scipy's minimize function and transform the Sharpe Ratio by multiplying it by a negative.  Thus, our new goal will be to minimize the new function $f$ subject to no short selling.  Results are presented below:

```python
f = lambda x,y,z: -sharpe(x, y, z)
cons = ({'type': 'eq', 'fun': lambda x:  sum(x)-1})
mvo_results = minimize(f, np.array([0.5, 0.5]), args=(ret, cov),
                       constraints=cons, bounds=((0.0, 1.0), (0.0, 1.0)))
iwt_wt, bnd_wt = mvo_results.x
print 'Maximum Sharpe Ratio: %.4f\n' % sharpe(mvo_results.x, ret, cov)
print 'MVO Portfolio Weights:\nIWV Wt: %0.3f\nBND Wt: %0.3f' % (iwt_wt, bnd_wt)
print 'Portfolio Return: %.4f' % port_avg_ret(mvo_results.x, ret)
print 'Portfolio Std. Deviation: %.4f' % port_sd(mvo_results.x, cov)
```
